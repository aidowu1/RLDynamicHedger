#### Set demo parameters

```python
SIMULATION_PATH_INDEX = 0
IS_PLOT_TO_SCREEN = True
HEDGING_TYPES = [HedgingType.gbm, HedgingType.sabr, HedgingType.heston]
USE_CASES = ["low_expiry", "high_expiry", "low_trading_cost", "high_trading_cost", "low_trading_freq",
             "high_trading_freq", "high_moneyness", "low_moneyness"]
# USE_CASES = ["high_trading_freq"]
```

#### Generate the comparative performance results

```python
def generate(
    hedging_type: HedgingType,
    model_use_case: str ="high_trading_cost",
    simulation_path_index: int = SIMULATION_PATH_INDEX,
    is_plot_to_screen: bool = True
):
    """
    Entry point to run the generation of RL comparative performance results for each hedging type
    :param hedging_type: Hedging type
    :return: None
    """
    print("Start of comparative RL agent result generations\n")
    
    
    print("Using parameter settings:")
    print(f"hedging_type: {hedging_type}")
    print(f"simulation path index: {simulation_path_index}")
    print("\nThis will take a few seconds to complete..\n")

    parameter_settings_data = Helpers.getParameterSettings(configs.DEFAULT_SETTINGS_NAME)
    parameters = Parameters(**parameter_settings_data)
    parameters.hedging_type = hedging_type
    run_scenario_parameters = getRunScenarioParams(parameters, scenario=model_use_case)
    print(f"Parameters used to run the simultion use case are:\n\n{run_scenario_parameters}\n\n")
    
    model = GenerateComparativePerformanceResults(
        hedging_type,
        simulation_path_index,
        is_plot_to_screen,
        model_use_case,
        parameters=run_scenario_parameters
    )
    result = model.execute()

```

#### Plot comparative performance metrics for all simulation use cases

```python
def runAllUseCases(
    use_cases: List[str]=USE_CASES, 
    hedging_types: List[HedgingType]=HEDGING_TYPES):
    """
    Generates the comprative results for all the simulation use cases
    :param use_cases: Use cases
    :param hedging_types: Hedging types
    """
    start_time = time.process_time()
    print(f"Start of RL agents comparative result generation cycles for 4 RL algorithms\n")
    for model_use_case in use_cases:
        for hedging_type in hedging_types:
            print(f"Plots for {model_use_case} use case and {hedging_type} simulation\n\n")
            generate(hedging_type=hedging_type, model_use_case=model_use_case)
        print(f"{LINE_DIVIDER}\n\n\n")
    end_time = time.process_time()
    elapsed_time_sec = round(end_time - start_time, 4)
    elapsed_time_min = round(elapsed_time_sec/60, 4)
    print(f"End of result generation for all the {len(use_cases)} uses cases for {len(hedging_types)}")
    print(f"Processing time was: {elapsed_time_sec} seconds | {elapsed_time_min} minutes")
    
```

```python
runAllUseCases()
```
